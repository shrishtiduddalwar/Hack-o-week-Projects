```python
import asyncio
import websockets
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import nest_asyncio

nest_asyncio.apply()

# ----- Create dataset -----
np.random.seed(0)

hours=np.random.randint(8,18,200)
temp=np.random.normal(30,3,200)
rain=np.random.choice([0,1],200)

load=40+hours*4+temp*1.5-rain*15+np.random.normal(0,8,200)

import pandas as pd
df=pd.DataFrame({"hour":hours,"temp":temp,"rain":rain,"load":load})

X=df[["hour","temp","rain"]]
y=df["load"]

model=LinearRegression()
model.fit(X,y)

# ----- WebSocket Server -----
async def predict(websocket):
    for i in range(20):

        sample={
            "hour":int(np.random.randint(8,18)),
            "temp":float(np.random.normal(30,3)),
            "rain":int(np.random.choice([0,1]))
        }

        X=[[sample["hour"],sample["temp"],sample["rain"]]]
        pred=model.predict(X)[0]

        sample["predicted_load"]=float(pred)

        await websocket.send(json.dumps(sample))
        await asyncio.sleep(1)

# ----- Client -----
loads=[]

async def client():
    uri="ws://localhost:8765"

    async with websockets.connect(uri) as ws:

        for i in range(20):

            msg=await ws.recv()
            data=json.loads(msg)

            loads.append(data["predicted_load"])

            plt.clf()
            plt.plot(loads)
            plt.title("Live Cafeteria Load Prediction")
            plt.pause(0.5)

            print(data)

# ----- Run server + client -----
async def main():

    server=await websockets.serve(predict,"localhost",8765)

    await client()

    server.close()
    await server.wait_closed()

await main()
```
