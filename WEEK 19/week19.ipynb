```python
users = {
    "admin1": "admin",
    "user1": "user"
}
```

```python
audit_logs = []

def log_access(user, action):
    audit_logs.append({
        "user": user,
        "role": users[user],
        "action": action
    })
```

```python
def access_data(user):
    role = users[user]

    if role == "admin":
        log_access(user, "Accessed full data")
        return " Full access granted"

    else:
        log_access(user, "Limited access")
        return " Limited access only"
```

```python
anomalies = [300, 200]
```

```python
def admin_panel(user):
    if users[user] != "admin":
        return " Access denied"

    print(" ADMIN PANEL\n")

    print(" Audit Logs:")
    for log in audit_logs:
        print(log)

    print("\n Anomaly Reports:")
    print(anomalies)
```

```python
print(access_data("user1"))
print(access_data("admin1"))

admin_panel("admin1")
```
